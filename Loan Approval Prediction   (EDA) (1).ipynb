### Data cleaning:

```python
### Check the columns in which missing values are present

print(App_info.isnull().sum())
```

#### Missing Values columns are 
1)Gender

2)Married

3)dependents

4)self_employed

5)Loanamount

6)loan_amount_term

7)credit_history

therefore first we fillup null/missing  values in above column

```python
#### fill missing values in categorical columns:
```

```python
print(App_info['Gender'].value_counts())
```

```python
## So we fillup Gender column 'Male'
App_info["Gender"].fillna("Male", inplace = True)
App_info['Gender'].isnull().sum()
```

```python
App_info['Gender'].unique()
```

```python
### Married column:
print(App_info['Married'].value_counts())
```

```python
App_info["Married"].fillna("Yes", inplace = True)
```

```python
###Dependants column:
print(App_info['Dependents'].value_counts())
```

```python
App_info["Dependents"].fillna("0", inplace = True)
```

```python
### Self_Employeed
print(App_info['Self_Employed'].value_counts())
```

```python
App_info["Self_Employed"].fillna("No", inplace = True)
```

```python
###Numerical columns:

####Loan amount
print(App_info['LoanAmount'].describe())
```

```python
### we fillup missing value by median 
App_info["LoanAmount"].fillna(128, inplace = True)
```

```python
#### Loan_Amount_Term
print(App_info['Loan_Amount_Term'].describe())
```

```python
App_info["Loan_Amount_Term"].fillna(360, inplace = True)
```

```python
###Credit_History
print(App_info['Credit_History'].value_counts())
```

```python
App_info['Credit_History'].dtype
```

```python
App_info["Credit_History"].fillna(1.0,inplace=True)
```

```python
App_info.to_csv("C:\\Loan prediction\\cleaned_data.csv")
```

### Data Visuallization:

```python
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
### Gender:
sns.countplot(App_info['Gender'])
plt.title('Gender')
plt.show()
```

#### from above plot we can say that,approximatly  75% male and 25% female in our data 

```python
sns.countplot(App_info['Loan_Status'])
```

```python
### In gender distribution how many of married:
sns.countplot('Gender',hue='Married',data=App_info)
```

#### from above countplot we can say that, more of applicants are Male and Married.

```python
### Is there association between Married and their Loan_status:
sns.countplot('Married',hue='Loan_Status',data=App_info)
plt.title('Married and Loan_status')
```

#### from above countplot we can say that, maximum number of applicant whose loan is approved are married people.

```python
sns.countplot('Credit_History',hue='Loan_Status',data=App_info)
plt.title('Credit_History and Loan_status')
```

#### from above plot we can say that, whose credit_history was good then loan was approved.

```python
### Is there assciation between Loan status and Loan_amount
sns.swarmplot(x='Loan_Status',y='LoanAmount',hue='Education',data=App_info)
plt.show()
```

#### from above swarmplot we can say that, there is association between loan amount and loan status and their education.

```python
#### Assciation between Loan status and Applicant income
sns.stripplot(x='Loan_Status',y='ApplicantIncome',hue='Property_Area',data=App_info)
plt.show()
```

```python
### Scatter plot : to check association between TWO numeric Columns/Variables
sns.scatterplot('ApplicantIncome','LoanAmount',data=App_info)
```

#### from above scatterplot we can conclude that, there is positive association between applicant income and their loan amount.  that is if income of applicant maximum then their more chances of getting higher loan amount.
